```python
import pandas as pd
import matplotlib.pyplot as plt

eventos = pd.read_excel('/content/eventos_rrhh2.xlsx')

# Filtrar solo accidentes
accident_subtypes = ['Sobreesfuerzo', 'Cada', 'Golpe/Contusin', 'Atrapamiento', 'Corte/Laceracin', 'Quemadura']
accidentes = eventos[eventos['subtipo_evento'].isin(accident_subtypes)]

# Contar cantidad de accidentes por rea
conteo = accidentes.groupby('area_empleado').size().sort_values(ascending=False)

# Grfico
conteo.plot(kind='bar')
plt.title('Cantidad de accidentes laborales por rea')
plt.ylabel('Cantidad de accidentes')
plt.xlabel('rea')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('accidentes_volumen_area.png', dpi=150)
plt.show()


#El ausentismo puede deberse o no a accidentes, indicado por la columna subtipo_evento con valores como 'Sobreesfuerzo', 'Cada', 'Golpe/Contusin'para verlo correctamente.
```
