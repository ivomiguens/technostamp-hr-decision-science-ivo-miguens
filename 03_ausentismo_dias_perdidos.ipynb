```python
#Ausentismo concentrado por rea - Identificar ausentismo por reas especficas

import pandas as pd
import matplotlib.pyplot as plt

eventos = pd.read_excel('/content/eventos_rrhh2.xlsx')

# Filtramos eventos con das perdidos

ausencias = eventos[eventos['dias_perdidos'] > 0]

# Agrupamos por rea

conteo = ausencias.groupby('area_empleado')['dias_perdidos'].sum().sort_values(ascending=False)

conteo.plot(kind='bar')
plt.title('Das perdidos por ausentismo por rea')
plt.ylabel('Total de das perdidos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('insight_3_ausentismo.png', dpi=150)
plt.show()
```
